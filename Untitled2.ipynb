## Introduction & Background **🥇**


Cookie Cats is a mobile puzzle game developed by Tactile Entertainment that has garnered a significant player base. Like many mobile games, it features gates, which are points where players must wait for some time or make an in-app purchase to continue playing. The placement of these gates can influence player engagement and retention, crucial metrics for the success of such games.

In this A/B test, the developers moved the first gate from level 30 to level 40 to study the effect on player behavior. The goal of the test is to measure if this change in gate position increases engagement and retention over a player's first 14 days after installation.

## *Problem*

The developers are faced with the challenge of determining the optimal gate placement in the game to maximize both engagement (measured by the number of game rounds played) and player retention (measured by whether players return on day 1 and day 7). Placing the gate too early could cause players to churn before they become invested, while placing it too late might result in lost revenue opportunities.

The dataset contains information about:


*   User ID: A unique identifier for each player.
*   Version: Either gate_30 (control group) or gate_40 (test group).
*   Game Rounds: The total number of rounds a player completed within the first 14 days.
*   Retention (Day 1): Whether the player returned on the day after installation.
*   Retention (Day 7): Whether the player returned after 7 days.

The goal is to use these metrics to identify the impact of moving the gate on player behavior.



## *Hypothesis*



- Null Hypothesis (H₀):  $\mu_{\text{gate_30}}$ = $\mu_{\text{gate_40}}$
  

- Alternative Hypothesis ( Hₐ): $\mu_{\text{gate_30}}$ > $\mu_{\text{gate_40}}$
